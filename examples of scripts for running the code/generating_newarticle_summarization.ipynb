```python

import io
from contextlib import redirect_stdout


def save_output_to_file(func, file_path, *args, **kwargs):
    """
    Execute a function, capture its printed output, 
    and save the output to a specified text file.
    
    Parameters:
    - func: The function to be executed.
    - file_path: The path to the text file where output will be saved.
    - *args: Positional arguments to pass to the function.
    - **kwargs: Keyword arguments to pass to the function.
    
    Returns:
    - The return value of the function.
    """
    # Create a StringIO object to capture the output
    output_buffer = io.StringIO()
    
    # Redirect stdout to the StringIO buffer and execute the function
    with redirect_stdout(output_buffer):
        return_value = func(*args, **kwargs)
    
    # Retrieve the captured output from the buffer
    console_output = output_buffer.getvalue()
    
    # Save the output to a text file
    with open(file_path, 'w') as file:
        file.write(console_output)
    
    return return_value
```
