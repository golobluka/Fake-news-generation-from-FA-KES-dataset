```python
import ollama
import os
import glob

prompt = """
You are a journalist working on world news. Extract {number_of_facts} types of facts from the given article.

Fact types should be of the form:
{{
  "name": "name of type of fact",
  "description": "What information does the fact contain.",
  "examples": "Some examples of the facts of prescribed type."
}}

OUTPUT EXAMPLES
{{
  "name": "Turnout",
  "description": "The number of people who take part in election.",
  "examples": "1000, 10k"
}},
{{
  "name": "Gender or age group",
  "description": " of casualty indicates if the casualties are male or female, or specify their age group.",
  "examples": "Male, Female, Child, Adult, Senior"
}},
{{
  "name": "Cause of death",
  "description": " specifies the weapons used by the aggressor (e.g., shooting, shelling, chemical weapons, etc.)",
  "examples": "Shooting, Shelling, Chemical weapons"
}},
{{
  "name": "Type",
  "description": " of casualty classifies the casualties as a civilian or non-civilian (e.g., military personnel are non-civilians).",
  "examples": "Civilian, Non-civilian"
}},
{{
  "name": "Allegations of Weapons Transfer",
  "description": "Hillary Clinton is accused of allowing American weapons to reach ISIS terrorists, via shipments initially intended for Libyan rebels.",
  "examples": "Weapons sent to Qatar, ISIS involvement, Assad opposition"
}},
{{
  "name": "Wikileaks Email Disclosure",
  "description": "Wikileaks released a significant email archive showing Hillary Clinton's alleged awareness of weapon flows to terrorists and State Department activities.",
  "examples": "1,700 emails, Julian Assange claims, Freedom of Information Act"
}},
{{
  "name": "Impact of Wikileaks Library",
  "description": "Wikileaks has released documents that have been used in legal cases and political events to hold organizations accountable and influence governments.",
  "examples": "CIA accountability, termination of government officials, influence on elections"
}}
END OF OUTPUT EXAMPLES

BEGINNING OF THE ARTICLE
{article}
END OF THE ARTICLE
"""

# Determine the path to the Fakes700 directory
parent_directory = os.path.dirname(os.getcwd())
fakes_directory = os.path.join(parent_directory, 'ACL2019_Data','Fakes700')

# Specify the input text file
filename = "article_0.txt"
text_files = os.path.join(fakes_directory, filename)

# Function to read and print contents of all text files (or process them)
def load_articles(file_paths):
    articles = []
    for file_path in file_paths:
        with open(file_path, 'r', encoding='utf-8') as file:
            content = file.read()
            articles.append(content)
            print(f"Loaded article from {file_path}:\n{content}\n")
            # Optionally, process the content further or store it in a data structure
    return articles

# Load articles from the listed text files
articles = load_articles([text_files])

response = ollama.chat(model="Llama3.1:8B", messages=[
    {
        'role': 'user',
        'content': prompt.format(article=articles[0], number_of_facts=4),
        'temperature': 0.2,
    }
])

output = response['message']['content']
print(output)
```
